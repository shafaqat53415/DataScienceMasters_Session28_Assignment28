# Assignment - 28: In this assignment students have to find the frequency of words in a webpage. User can use urllib and BeautifulSoup to extract text from webpage.

```python
from bs4 import BeautifulSoup
import urllib.request
import nltk

response = urllib.request.urlopen('http://php.net/')
html = response.read()
soup = BeautifulSoup(html,"html5lib")
```

```python
soup
```

```python
text = soup.get_text(strip=True)
```

```python
words = nltk.tokenize.word_tokenize(text)
```

```python
words
```

```python
fd = nltk.FreqDist(words)
```

```python
fd
```

    It can be observed that there are punctuations in the above tokens. Let us try to split the string.

```python
words = [t for t in text.split()]
```

```python
words
```

```python
fd_ = nltk.FreqDist(words)
```
